```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

date_index = pd.date_range(start='2024-04-01', periods=15, freq='D')
random_values = np.random.randint(500, 1001, size=15)  #randint принимает верхнюю границу не включительно, поэтому 1001
date_series = pd.Series(random_values, index=date_index)

#  Построение графика и отображение тренда

plt.figure(figsize=(12, 6))

#  график значений
plt.plot(date_series, label='Значения')

#  среднее значение
mean_value = date_series.mean()

#  пунктирнуя линия среднего значения
plt.axhline(mean_value, color='red', linestyle='--', label=f'Среднее: {mean_value:.2f}') #Добавляем горизонтальную линию.  f-строка для форматирования среднего

plt.title("Изменение случайных значений в апреле 2024", fontsize=16)
plt.xlabel("Дата", fontsize=12)
plt.ylabel("Значение", fontsize=12)

#  сетка
plt.grid(True)

plt.xticks(rotation=45)
plt.tight_layout()

plt.legend()

plt.show()
```
