```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

date_index = pd.date_range(start='2024-04-01', periods=15, freq='D')
random_values = np.random.randint(500, 1001, size=15)  #randint     ,  1001
date_series = pd.Series(random_values, index=date_index)

#      

plt.figure(figsize=(12, 6))

#   
plt.plot(date_series, label='')

#   
mean_value = date_series.mean()

#     
plt.axhline(mean_value, color='red', linestyle='--', label=f': {mean_value:.2f}') #  .  f-   

plt.title("     2024", fontsize=16)
plt.xlabel("", fontsize=12)
plt.ylabel("", fontsize=12)

#  
plt.grid(True)

plt.xticks(rotation=45)
plt.tight_layout()

plt.legend()

plt.show()
```
