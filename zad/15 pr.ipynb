```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

date_index = pd.date_range(start='2024-03-01', periods=30, freq='D')
random_values = np.random.randint(50, 151, size=30)  # randint принимает верхнюю границу не включительно
date_series = pd.Series(random_values, index=date_index)

#  график
plt.figure(figsize=(12, 6))  
plt.plot(date_series) 

plt.title("Изменение случайных значений в марте 2024", fontsize=16)
plt.xlabel("Дата", fontsize=12)
plt.ylabel("Значение", fontsize=12)

#  сетка
plt.grid(True)

plt.xticks(rotation=45) 
plt.tight_layout()  

plt.show()
```
